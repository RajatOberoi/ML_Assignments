```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv')
df = pd.DataFrame(data)
df.head()
```

```python
"""
Q) Plot atleast 20 visualization for the same dataset with different data points.
   Ensure that the type of visualization is appropriate and follows this chart .

"""
```

```python
#1)By default kde=True,hist=True
sns.distplot(df['MinTemp'])
```

```python
#2)Kernel Density Estimation 1D plot
sns.distplot(df['MinTemp'],hist=False,rug=True)
```

```python
#3) Histogram
sns.distplot(df['MaxTemp'],kde=False,rug=True)
```

```python
#4) Kernel Density Estimation 1d plot
sns.kdeplot(df['Rainfall'],shade=True)
```

```python
#5)
sns.kdeplot(df['MaxTemp'],label='original')
sns.kdeplot(df['MaxTemp'],bw=1)
sns.kdeplot(df['MaxTemp'],bw=2,label='bw=2')
plt.legend()
```

```python
#6) Hexbin Plot
sns.jointplot(kind='hex',y=df['Rainfall'],x=df['Sunshine'])
```

```python
#7)Scatter
sns.jointplot(kind='scatter',y=df['MaxTemp'],x=df['MinTemp'])
```

```python
#8) Kernel Density Estimation 2D plot
sns.jointplot(kind = 'kde',y=df['MaxTemp'],x=df['MinTemp'])
```

```python
#9) Pair plot
x = pd.DataFrame(df,columns=['MaxTemp','MinTemp','Cloud9am','Cloud3pm'])
sns.pairplot(x)
```

```python
#10)
sns.pairplot(x,hue='MinTemp')
```

```python
#11)
sns.stripplot(x='Temp9am',y='Temp3pm',data=df,jitter=True)
```

```python
#12)Strip plot
sns.stripplot(x='Temp9am',y='Temp3pm',data=df)
```

```python
#13)Swarm Plot
sns.swarmplot(x='Cloud9am',y='Cloud3pm',data=df)
```

```python

#14)Box Plot
sns.boxplot(x='Cloud9am',y='Cloud3pm',data=df)
```

```python
#15)
sns.boxplot(x='RainToday',y='Humidity3pm',data=df,hue='WindDir3pm',dodge=True)
```

```python
#16)
sns.barplot(x='Cloud9am',y='Cloud3pm',data=df)
```

```python
#17)
sns.boxplot(x='RainToday',y='Humidity9am',data=df,hue='WindDir3pm',dodge=False)
```

```python
#18)Count Plot
sns.countplot(x='Cloud3pm',data=df)
```

```python
#19)
sns.barplot(x='RainToday',y='Rainfall',hue='Humidity3pm',data=df)
```

```python

#20)
sns.countplot(x='WindDir9am',data=df)
```
